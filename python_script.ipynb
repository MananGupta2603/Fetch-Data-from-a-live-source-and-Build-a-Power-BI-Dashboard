```python

import pandas as pd
uri= "https://raw.githubusercontent.com/LokeshKumarChauhan/DE_with_powerBI/main/Walmart.csv"
df=pd.read_csv(uri)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
def parse_dates(date):
    try:
        return pd.to_datetime(date, format='%d-%m-%Y')
    except ValueError:
        return pd.to_datetime(date, format='%d/%m/%Y')


df['Order Date'] = df['Order Date'].apply(parse_dates).dt.strftime('%d-%m-%Y')
df['Ship Date'] = df['Ship Date'].apply(parse_dates).dt.strftime('%d-%m-%Y')
df
```
